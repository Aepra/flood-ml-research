```python
import pandas as pd
from pathlib import Path

# Get project root path
try:
    PROJECT_ROOT = Path(__file__).parent.parent
except NameError:
    cwd = Path.cwd()
    PROJECT_ROOT = cwd.parent if cwd.name == "notebooks" else cwd

# Load grid
grid_path = PROJECT_ROOT / "data" / "processed" / "makassar_grid_250m.parquet"
grid = pd.read_parquet(grid_path)

print(f"GRID LOADED: {len(grid)}")
print(f"From: {grid_path}")
grid.head()
```

```python
import pandas as pd

start_date = "2018-01-01"
end_date   = "2022-12-31"

dates = pd.date_range(start=start_date, end=end_date, freq="D")

print("TOTAL DAYS:", len(dates))
```

```python
total_rows = len(grid) * len(dates)

print("GRID:", len(grid))
print("DAYS:", len(dates))
print("TOTAL SPATIO-TEMPORAL ROWS:", total_rows)
```

```python
import os
import pandas as pd
import itertools
from pathlib import Path

# Get project root path
try:
    PROJECT_ROOT = Path(__file__).parent.parent
except NameError:
    cwd = Path.cwd()
    PROJECT_ROOT = cwd.parent if cwd.name == "notebooks" else cwd

data_processed = PROJECT_ROOT / "data" / "processed"
data_processed.mkdir(parents=True, exist_ok=True)

grid_ids = grid["grid_id"].values
dates_list = dates

pairs = list(itertools.product(grid_ids, dates_list))

df = pd.DataFrame(pairs, columns=["grid_id", "date"])

print("FINAL SHAPE:", df.shape)
df.head()
```

```python
from pathlib import Path

# Get project root path
try:
    PROJECT_ROOT = Path(__file__).parent.parent
except NameError:
    cwd = Path.cwd()
    PROJECT_ROOT = cwd.parent if cwd.name == "notebooks" else cwd

data_processed = PROJECT_ROOT / "data" / "processed"
parquet_path = data_processed / "makassar_spatiotemporal_base.parquet"

df.to_parquet(parquet_path, index=False)

print(f"✔ PARQUET SAVED -> {parquet_path}")
print(f"   File size: {parquet_path.stat().st_size / 1e6:.2f} MB")
```
